```python
from inaregParser import RegParser
import json
```

```python
test_UU = RegParser(r"../data/UU Nomor 6 Tahun 2023 PENGGANTI UNDANG-UNDANG.pdf", parse_now=True)
```

```python
print("Judul:", test_UU.title)
print("\n")
print("Info:", test_UU.info())
```

```python
print("Konsideran Menimbang:", test_UU.get_philosophical_consideration())
```

```python
# Ambil hasil parsing dalam bentuk dictionary
parsed_data = test_UU.to_json(return_dict=True)

# Cetak ke layar (opsional)
# print(json.dumps(parsed_data, indent=2, ensure_ascii=False))
print(parsed_data)
# Simpan ke file JSON
with open("output_uu13.json", "w", encoding="utf-8") as f:
    json.dump(parsed_data, f, indent=2, ensure_ascii=False)

print("✅ File berhasil disimpan ke output_uu13.json")
```
